The code is not robust, meanning if there is a conflict with no existing folder or existing file with the same name there is no garantee.

```python
#Making a proper folder structure
def new_folder(mypath):
    if not os.path.isdir(mypath):
        os.makedirs(mypath)
        
new_folder('images\\train')
new_folder('images\\valid')
new_folder('images\\test')
```

```python
#df_image = pd.read_csv("validation-images.csv")           # Image name and url
#df_label = pd.read_csv("validation-annotations-bbox.csv") # ImageIDs, labels & TagInfos
df_image = pd.read_csv("train-images-boxable.csv")           # Image name and url
df_label = pd.read_csv("train-annotations-bbox.csv") # ImageIDs, labels & TagInfos
df_desc  = pd.read_csv("class-descriptions.csv", header=None)          # Class definitions and decoded tags
df_desc.columns = ["ID", "class"]
```

# Chose your categories

```python
print('You can choose from this labels (' + str(df_desc.size) + "):\n")
lst = df_desc['class'].tolist()
lst_sorted = sorted(lst)
for i in lst_sorted:
    print(i)
```

```python
example_1 = "Antelope"
example_2 = "Zebra"
train_split = 400
valid_split = 100
test_split  = 100
#folder = str('C:\\0_STORAGE\\0_DeepLearning\\jupiter\\images\\')
folder = str('C:\\0_STORAGE\\0_DeepLearning\\jupiter\\KisHazi\\04_OpenImageTransferLearn\\images\\')
```

```python
# Categorie chack if it is an existing option from the list
if example_1 in lst_sorted:
    print("Excellent choise. "+str(example_1)+" it is!")
    new_folder('images\\train\\'+str(example_1))
    new_folder('images\\valid\\'+str(example_1))
    new_folder('images\\test\\'+str(example_1))
else:
    print("You should chose another class instead of " + str(example_1))
if example_2 in lst_sorted:
    print("Excellent choise. "+str(example_2)+" it is!")
    new_folder('images\\train\\'+str(example_2))
    new_folder('images\\valid\\'+str(example_2))
    new_folder('images\\test\\'+str(example_2))
else:
    print("You should chose another class instead of " + str(example_2))
```

```python
ID1 = df_desc.loc[df_desc['class'] == example_1].iloc[0].loc['ID']
ID2 = df_desc.loc[df_desc['class'] == example_2].iloc[0].loc['ID']
```

```python
def download_image(url, name, folder):
    fullname = str(name)+".jpg"
    urllib.request.urlretrieve(url,folder+fullname) 
```

```python
df_ex1 = df_label[df_label['LabelName'] == ID1]
df_ex2 = df_label[df_label['LabelName'] == ID2]

# Drop out the duplons
df_ex1 = df_ex1.drop_duplicates(subset='ImageID')
df_ex2 = df_ex2.drop_duplicates(subset='ImageID')

# Parsing the right amount of data
df_ex1_train = df_ex1.iloc[0:train_split, [0]]
df_ex2_train = df_ex2.iloc[0:train_split, [0]]
df_ex1_valid = df_ex1.iloc[train_split:(train_split+valid_split), [0]]
df_ex2_valid = df_ex2.iloc[train_split:(train_split+valid_split), [0]]
df_ex1_test = df_ex1.iloc[(train_split+valid_split):(train_split+valid_split+test_split), [0]]
df_ex2_test = df_ex2.iloc[(train_split+valid_split):(train_split+valid_split+test_split), [0]]
```

```python
print('df_ex1.size: ' + str(df_ex1.size))
print('df_ex2.size: ' + str(df_ex1.size))
print('df_ex1_train.size: ' + str(df_ex1_train.size))
print('df_ex2_train.size: ' + str(df_ex2_train.size))
print('df_ex1_valid.size: ' + str(df_ex1_valid.size))
print('df_ex2_valid.size: ' + str(df_ex2_valid.size))
print('df_ex1_test.size: ' + str(df_ex1_test.size))
print('df_ex2_test.size: ' + str(df_ex2_test.size))
```

```python
print('Downloading...')
index_1 = 0
index_2 = 0

print('=================TRAIN=================')
for id in df_ex1_train['ImageID']:
    pic = example_1+str(index_1)
    name = str(id) + '.jpg'
    print(pic)
    url = df_image[df_image['image_name'] == name].iloc[0].loc['image_url']
    print(url)
    download_image(url, pic, folder+str('train\\')+example_1+str('\\'))
    index_1 = index_1+1
      
for id in df_ex2_train['ImageID']:
    pic = example_2+str(index_2)
    name = str(id) + '.jpg'
    print(pic)
    url = df_image[df_image['image_name'] == name].iloc[0].loc['image_url']
    print(url)
    download_image(url, pic, folder+str('train\\')+example_2+str('\\'))
    index_2 = index_2+1      
    
print('=================VALID=================')    
for id in df_ex1_valid['ImageID']:
    pic = example_1+str(index_1)
    name = str(id) + '.jpg'
    print(pic)
    url = df_image[df_image['image_name'] == name].iloc[0].loc['image_url']
    print(url)
    download_image(url, pic, folder+str('valid\\')+example_1+str('\\'))
    index_1 = index_1+1

for id in df_ex2_valid['ImageID']:
    pic = example_2+str(index_2)
    name = str(id) + '.jpg'
    print(pic)
    url = df_image[df_image['image_name'] == name].iloc[0].loc['image_url']
    print(url)
    download_image(url, pic, folder+str('valid\\')+example_2+str('\\'))
    index_2 = index_2+1
      
print('=================TEST=================')      
for id in df_ex1_test['ImageID']:
    pic = example_1+str(index_1)
    name = str(id) + '.jpg'
    print(pic)
    url = df_image[df_image['image_name'] == name].iloc[0].loc['image_url']
    print(url)
    download_image(url, pic, folder+str('test\\')+example_1+str('\\'))
    index_1 = index_1+1
      
for id in df_ex2_test['ImageID']:
    pic = example_2+str(index_2)
    name = str(id) + '.jpg'
    print(pic)
    url = df_image[df_image['image_name'] == name].iloc[0].loc['image_url']
    print(url)
    download_image(url, pic, folder+str('test\\')+example_2+str('\\'))
    index_2 = index_2+1

print('DOWNLOADING COMPLETE')
```
